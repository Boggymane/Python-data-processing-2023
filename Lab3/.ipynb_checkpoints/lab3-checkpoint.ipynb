```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1.
df = pd.read_excel('En_In.xls', skiprows=17, skipfooter=38)
df = df.iloc[:, 2:]
df.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
df.index = np.arange(1, len(df) + 1)
pd.set_option('display.float_format', '{:.3f}'.format)
```

```python
#2.Конвертація значень в гДж та обробка відсутніх даних
df['Energy Supply'] = pd.to_numeric(df['Energy Supply'], errors='coerce') * 1e6
columns_to_convert = ['Energy Supply', 'Energy Supply per Capita', '% Renewable']
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
#3. Перейменуйте наступний список країн
country_mapping = {
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
}

df['Country'] = df['Country'].replace(country_mapping)
```

```python
#4.Видалення цифр та дужок
df['Country'] = df['Country'].str.replace(r'(\d+|\([^)]*\))', '', regex=True).str.strip()
```

```python
#Вивід для завдань 1-4
df.loc[df['Country'].isin(['American Samoa', 'South Korea', 'Bolivia' ])]
```

```python
#5.
gpd = pd.read_csv('gpd.csv', skiprows=4)
country_mapping_gpd = {
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
}
gpd['Country Name'] = gpd['Country Name'].replace(country_mapping_gpd)
gpd.head(1)
```

```python
#6.
scimago = pd.read_excel('scimagojr.xlsx')
scimago
```

```python
#7.
merged_df = pd.merge(pd.merge(df, gpd, left_on='Country', right_on='Country Name'), scimago, how='inner', left_on='Country', right_on='Country')
selected_columns = ['Country', 'Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',
                    'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable',
                    '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
result = merged_df[selected_columns].set_index('Country').sort_values(by='Rank').head(15)
result.head(3)
```

```python
result.shape
```

```python
#8.
def avergdp():
    gdp_columns = [str(year) for year in range (2006, 2016)]
    return result[gdp_columns].mean(axis=1).sort_values(ascending=False)
avergdp()
```

```python
#9.
def gdp_change5():
    fifth_country = avergdp().index[4]
    return fifth_country, result.loc[fifth_country]['2015'] - result.loc[fifth_country]['2006']

gdp_change5()
```

```python
#10.
def max_renewable(dataframe):
    country = dataframe['% Renewable'].idxmax()
    return country, dataframe.loc[country]['% Renewable']
    
max = max_renewable(result)
max
```

```python
#11.
def population(data):
    data['Population'] = data['Energy Supply'] / data['Energy Supply per Capita']
    sixth_country = data.sort_values(by='Population', ascending=False).iloc[5]
    return sixth_country.name, sixth_country['Population']

popu = population(result)
popu
```

```python
#12.
def citper(data):
    data['Citations per Capita'] = data['Citable documents'] / (data['Energy Supply'] / data['Energy Supply per Capita'])
    corr = data['Citations per Capita'].corr(data['Energy Supply per Capita'])
    return corr

corr = citper(result)
print(corr)
```

```python
#13.
def task_thirteen():
    median = result['% Renewable'].median()
    result['HighRenew'] = (result['% Renewable'] >= median).astype(int)
    return result['HighRenew']

task_thirteen()
```

```python
#14.
result_df = result.copy()

ContinentDict = {'China': 'Asia', 'United States': 'North America', 'Japan': 'Asia', 'United Kingdom': 'Europe',
                 'Russian Federation': 'Europe', 'Canada': 'North America', 'Germany': 'Europe', 'India': 'Asia',
                 'France': 'Europe', 'South Korea': 'Asia', 'Italy': 'Europe', 'Spain': 'Europe', 'Iran': 'Asia',
                 'Australia': 'Australia', 'Brazil': 'South America'}

result_df['Continent'] = result_df.index.map(ContinentDict)

continent_stats = result_df.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])

continent_stats.columns = ['size', 'sum', 'mean', 'std']

continent_stats
```

```python
#15.
bubble = result.copy()

ContinentDict = {'China': 'Asia', 'United States': 'North America', 'Japan': 'Asia', 'United Kingdom': 'Europe',
                 'Russian Federation': 'Europe', 'Canada': 'North America', 'Germany': 'Europe', 'India': 'Asia',
                 'France': 'Europe', 'South Korea': 'Asia', 'Italy': 'Europe', 'Spain': 'Europe', 'Iran': 'Asia',
                 'Australia': 'Australia', 'Brazil': 'South America'}
bubble['Continent'] = bubble.index.map(ContinentDict)
colors_dict = {'Asia': 'red', 'North America': 'blue', 'Europe': 'green', 'Australia': 'yellow', 'South America': 'orange'}
bubble['Color'] = bubble['Continent'].map(colors_dict)

plt = bubble.plot(x='Rank', y='% Renewable', kind='scatter', xticks=range(1, 16),
                  s=6 * bubble['2014'] / 10**10, alpha=0.75, figsize=[16, 5], c=bubble['Color'])

for i, txt in enumerate(bubble.index):
    plt.annotate(txt, (bubble['Rank'].iloc[i], bubble['% Renewable'].iloc[i]), fontsize=9, ha='center')
```
